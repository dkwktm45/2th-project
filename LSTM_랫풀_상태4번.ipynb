```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# 결측치 처리하는 부분도 코드로 넣어버림
```

```python
json_list = []
for i in ['17', '19', '20', '21', '22']:
    for j in range(1, 10):
        for k in range(697, 817):
            try:
                file = '../../0.jsondata/Day'+i+'_2010'+i+'_F/D'+i+'-'+str(j)+'-'+str(k)+'.json'
                print(file)
                with open(file, 'r', encoding='utf-8') as f:
                    jsondata = json.load(f)
                    if jsondata['type_info']['exercise'] == '랫풀 다운':
                        if len(jsondata['frames']) == 16 :
                                json_list.append(jsondata)
            except FileNotFoundError as n:
                print(n)
                continue
```

```python
land_name_list = list(json_list[0]['frames'][0]['view1']['pts'])
```

```python
json_list[0]['type_info']['conditions']
```

# 네번째운동상태 :  수축 시 몸통-팔꿈치 사이 모아줌

```python
json_list[0]['type_info']['conditions'][3]
```

```python
json_list[0]['type_info']['conditions'][3]['value']
# 두번째 0 이 첫번째 운동상태를 의미
```

```python
land_list = []
frame_list = []
X_list = []
y_list = []

for jsondata in json_list:
    frames = jsondata['frames']
    for i in range(1,6):
        frame_list = []
        for frame in frames:
            pts = frame['view'+str(i)]['pts']
            land_list = []
            for land_name in land_name_list:
                x = pts[land_name]['x']
                y = pts[land_name]['y']
                land_list.append(x)
                land_list.append(y)
            frame_list.append(land_list)
        X_list.append(frame_list)
        
    con = jsondata['type_info']['conditions'][3]['value']
    for i in range(5):
        y_list.append(con)
```

```python
len(X_list)
```

```python
len(y_list)
```

```python
X = np.array(X_list)
y = np.array(y_list)
```

```python
X.shape
```

```python
y.shape
```

```python
y_one_hot = pd.get_dummies(y)
```

```python
y_one_hot.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y_one_hot,
                                                    random_state=0,
                                                    test_size=0.2
                                                   )
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding, Activation
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, SimpleRNN, Dropout, TimeDistributed, Conv2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import TensorBoard
```

```python
model = Sequential()
model.add(LSTM(256, return_sequences=True, activation='tanh', input_shape=(16,48)))
model.add(LSTM(128, return_sequences=True, activation='tanh'))
model.add(LSTM(64, return_sequences=False, activation='tanh'))
model.add(Dense(64, activation='tanh'))
model.add(Dense(128, activation='tanh'))
model.add(Dense(64, activation='tanh'))
model.add(Dense(32, activation='tanh'))
model.add(Dense(16, activation='tanh'))
model.add(Dense(2, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.0003),
             metrics=['categorical_accuracy']) 
#Adam(learning_rate=0.0003)
```

```python
#          validation_split=0.3,
```

```python
h = model.fit(X_train, y_train, validation_split=0.3,
          epochs=500, batch_size=100)
```

```python
# 시각화
plt.figure(figsize=(15, 5))

# train 데이터
plt.plot(h.history['categorical_accuracy'],
        label='acc'
        )
# val 데이터
plt.plot(h.history['val_categorical_accuracy'],
        label='val_acc'
        )
plt.legend()
plt.show()
```

```python
eva = model.evaluate(X_test, y_test)
eva
```
